## Plotting

```python
fig, ax = plt.subplots()
prova[['temp_cpu', 'temp_h', 'temp_p', 'humidity']].plot(ax=ax, figsize=(16, 5))
ax.legend(['temp_cpu', 'temp_h', 'temp_p', 'humidity'])
fig, ax = plt.subplots()
prova[['pressure']].plot(ax=ax, figsize=(16, 5))
ax.legend(['pressure'])
```

```python
for metrics in [['pitch', 'roll', 'yaw'],
       ['mag_x', 'mag_y', 'mag_z'],
       ['accel_x', 'accel_y', 'accel_z'],
       ['gyro_x', 'gyro_y', 'gyro_z']]:
       legend = []
       for m in metrics:
              legend.append(m)
       fig, ax = plt.subplots()
       prova[metrics].plot(ax=ax, figsize=(16, 5))
       ax.legend(legend)
```

```python
colors = ['orange', 'blue', 'green', 'red']
c = 0
for metrics in ['pitch', 'roll', 'yaw', 'mag_x', 'mag_y', 'mag_z']:
       legend = []
       legend.append(metrics)
       fig, ax = plt.subplots()
       prova[metrics].plot(ax=ax, figsize=(16, 5), color=colors[c])
       ax.legend(legend)
       c = c + 1
       if c == 4:
              c = 0
```

```python
for metrics in [['compass', 'pitch', 'roll', 'yaw'],
       ['magnetometer', 'mag_x', 'mag_y', 'mag_z'],
       ['accelerometer', 'accel_x', 'accel_y', 'accel_z'],
       ['gyroscope', 'gyro_x', 'gyro_y', 'gyro_z']]:
       fig = plt.figure(figsize=(10, 5))
       ax = plt.axes(projection='3d')
       ax.scatter3D(prova[metrics[1]], prova[metrics[2]], prova[metrics[3]], c=prova[metrics[3]], cmap='viridis');
       plt.title(metrics[0])
```
