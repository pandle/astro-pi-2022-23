```python
import warnings
warnings.filterwarnings('ignore')

%matplotlib widget
import matplotlib.pyplot as plt

import pandas as pd
prova = pd.read_csv('AstroPi2021/Juno/juno_data.csv')
```

## Plotting

```python
fig, ax = plt.subplots()
prova[['____CPU_temp', '__Sense_temp', '___Sense_Hum']].plot(ax=ax, figsize=(16, 5))
ax.legend(['CPU_temp', 'Sense_temp', 'Sense_Hum'])
```

```python
for metrics in [['_______pitch', '________roll', '_________yaw'],
       ['________accX', '________accY', '________accZ'],
       ['________magX', '________magY', '________magZ'],
       ['_______gyroX', '_______gyroY', '_______gyroZ']]:
       legend = []
       for m in metrics:
              legend.append(m.replace('_',''))
       fig, ax = plt.subplots()
       prova[metrics].plot(ax=ax, figsize=(16, 5))
       ax.legend(legend)
```

```python
colors = ['orange', 'blue', 'green', 'red']
c = 0
for metrics in ['_______pitch', '________roll', '_________yaw']:
       legend = []
       legend.append(metrics.replace('_',''))
       fig, ax = plt.subplots()
       prova[metrics].plot(ax=ax, figsize=(16, 5), color=colors[c])
       ax.legend(legend)
       c = c + 1
       if c == 4:
              c = 0
```

```python
for metrics in [['compass', '_______pitch', '________roll', '_________yaw'],
       ['accelerometer', '________accX', '________accY', '________accZ'],
       ['magnetometer', '________magX', '________magY', '________magZ'],
       ['gyroscope', '_______gyroX', '_______gyroY', '_______gyroZ']]:
       fig = plt.figure(figsize=(10, 5))
       ax = plt.axes(projection='3d')
       ax.scatter3D(prova[metrics[1]], prova[metrics[2]], prova[metrics[3]], c=prova[metrics[3]], cmap='viridis');
       plt.title(metrics[0])
```
